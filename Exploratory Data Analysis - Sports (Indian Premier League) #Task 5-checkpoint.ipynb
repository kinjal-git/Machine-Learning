```python
season_data = df_matches[['id','season','winner']]
complete_data = df_delivery.merge(season_data, how = 'inner', left_on = 'match_id', right_on = 'id')
complete_data.head()
```

```python
#Exploring Matches dataset

df_matches.columns.values
```

```python
#Dropping unwanted columns

df_matches = df_matches.drop(columns = 'umpire3', axis = 1)
df_matches.head()
```

```python
#Wins per season

team_wins = df_matches.groupby('season')['winner'].value_counts()
team_wins
```

```python
#Exploring matches played per season

import seaborn as sns
plt.figure(figsize = (20,10))
sns.countplot('season', data = df_matches, palette = 'viridis')
plt.title('Number of Matches played in each IPL season', fontsize = 25)
plt.xlabel('season', fontsize = 20)
plt.ylabel('Matches', fontsize = 20)
plt.show()
```

```python
#Exploring number of matches won by a team

plt.figure(figsize = (18,10))
sns.countplot('winner', data = df_matches, palette = 'flare')
plt.title('Number of Matches won by a team', fontsize = 25)
plt.xticks(rotation = 60)
plt.xlabel('Teams', fontsize = 20)
plt.ylabel('Total Wins', fontsize = 20)
plt.show()
```

```python
df_matches['win_by'] = np.where(df_matches['win_by_runs']>0, 'Bat first', 'Bowl first')
```

```python
#Exploring wins by batting first or bowling first

Win = df_matches.win_by.value_counts()
colors = ['darkorange', 'limegreen']
plt.figure(figsize = (15,10))
plt.pie(x = Win.values, labels = np.array(Win.index), colors = colors, autopct='%1.1f%%', shadow=True, startangle = 90)
plt.title('Match Result', fontsize =  20)
plt.axis('equal')
plt.show()
```

```python
#Comparing matches won by batting or bowling first

plt.figure(figsize = (20,10))
sns.countplot('season', hue = 'win_by', data = df_matches, palette = 'cubehelix')
plt.title('Numbers of matches won by batting and bowling first ', fontsize = 25)
plt.xlabel('Season', fontsize = 20)
plt.ylabel('Count', fontsize = 20)
plt.show()
```

```python
#Plotting the toss decisions

Toss = df_matches.toss_decision.value_counts()
colors = ['darkorange', 'limegreen']
plt.figure(figsize = (10,10))
plt.pie(x = Toss.values, labels = np.array(Toss.index), colors = colors, autopct='%1.1f%%', shadow=True, startangle = 90)
plt.title('Toss Result', fontsize =  20)
plt.axis('equal')
plt.show()
```

```python
#Plotting graphs on the number of matches won by toss results

plt.figure(figsize = (20,10))
sns.countplot('season', hue = 'toss_decision', data = df_matches, palette = 'dark:salmon_r')
plt.title('Numbers of matches won by Toss Results ', fontsize = 25)
plt.xlabel('Season', fontsize = 20)
plt.ylabel('Count', fontsize = 20)
plt.show()
```

```python
#We will print the winners season wise

final_matches = df_matches.drop_duplicates(subset = ['season'], keep = 'last')
final_matches[['season','winner']].reset_index(drop = True).sort_values('season')
```

```python
#We will plot the graph of the percentage of Winning of teams in final matches

match = final_matches.win_by.value_counts()
colors = ['gold', 'purple']
plt.figure(figsize = (10,10))
plt.pie(x = match.values, labels = np.array(match.index), colors = colors, autopct = '%1.1f%%', shadow = True,startangle = 90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

```python
#We will plot the toss decisions in final matches

Toss = final_matches.toss_decision.value_counts()
colors = ['gold','purple']
plt.figure(figsize = (10,10))
plt.pie(x = Toss.values, labels = np.array(Toss.index), colors = colors, autopct = '%1.1f%%', shadow = True,startangle = 90)
plt.title('Toss Result',fontsize = 20)
plt.axis('equal')
plt.show()
```

```python
#We will print the name of top player in IPL

plt.figure(figsize = (20,10))
top_players = df_matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient = 'v', palette = 'hsv');
plt.show()
```

```python
#Exploring the most fours hit by team

fours = complete_data[complete_data['batsman_runs'] == 4]
fours.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
#Plotting the fours hit by players

fours_by_batsman = fours.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax = fours_by_batsman.iloc[:10,:].plot('batsman','four',kind = 'bar',color = 'magenta')
plt.title("Numbers of fours hit by players ",fontsize = 25)
plt.xticks(rotation = 50)
plt.xlabel("Player name",fontsize = 20)
plt.ylabel("No of fours",fontsize = 20)
plt.show()
```

```python
#Plotting the fours hit in each season

ax = fours.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind = 'bar',color = 'green')
plt.title("Numbers of fours hit in each season ",fontsize = 25)
plt.xticks(rotation = 50)
plt.xlabel("season",fontsize = 20)
plt.ylabel("No of fours",fontsize = 20)
plt.show()
```

```python
#Exploring the most sixes hit by teams

sixes = complete_data[complete_data['batsman_runs'] == 6]
sixes.groupby('batting_team')['batsman_runs'].agg([('runs by sixes','sum'),('sixes','count')])
```

```python
#Plotting the sixes hit by players

sixes_by_batsman = sixes.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax = sixes_by_batsman.iloc[:10,:].plot('batsman','six',kind = 'bar',color = 'blue')
plt.title("Numbers of sixes hit by players ",fontsize = 25)
plt.xticks(rotation = 50)
plt.xlabel("Player name",fontsize = 20)
plt.ylabel("No of sixes",fontsize = 20)
plt.show()
```

```python
#Plotting the sixes hit in each season

ax = sixes.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind = 'bar',color = 'red')
plt.title('Numbers of sixes hit in each season',fontsize = 25)
plt.xticks(rotation = 50)
plt.xlabel('season',fontsize = 20)
plt.ylabel('No of sixes',fontsize = 20)
plt.show()
```

```python
#Exploring the top 10 leading run scorer in IPL

batsman_score = df_delivery.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score = batsman_score.rename(columns = {'sum':'batsman_runs'})
print('Top 10 Leading Run Scorer in IPL')
batsman_score.iloc[:10,:]
```

```python
#Plotting dismissals in IPL

plt.figure(figsize = (20,10))
ax = sns.countplot(df_delivery.dismissal_kind)
plt.title('Dismissals in IPL',fontsize = 25)
plt.xlabel('Dismissals kind',fontsize = 20)
plt.ylabel('count',fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Exploring the most number of wicket takers

Dismissal = df_delivery.dropna(subset=['dismissal_kind'])
Dismissal = Dismissal[~Dismissal['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
Dismissal.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```
