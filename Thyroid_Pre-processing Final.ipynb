```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hnames=['age','sex','on thyroxine','query on thyroxine','on antithyroid medication','sick','pregnant','thyroid surgery','1131 treatment','query hypothyroid','query hyperthyroid','lithium','goitre','tumor','hypopituitary','psych','TSH measured','TSH','T3 measured','T3','TT4 measured','TT4','T4U measured','T4U','FTI measured','FTI','TBG measured','TBG','referral source','Diagnosis']
df = pd.read_csv("C:\\Users\\Admin\\Downloads\\thyroid0387.csv", names = hnames)
```

## Replace '?' by NaN

```python
df = df.replace('?', np.nan)
df = df.astype({'TSH' : 'float', 'T3' : 'float', 'TT4' : 'float', 'T4U' : 'float', 'FTI' : 'float', 'TBG' : 'float'})
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['TBG measured', 'TBG', 'referral source'], axis = 1)
df
```

# Replacing sex by int values  

```python
df.sex.replace({'F':1,'M':0},inplace=True)
round_Values = round(df.sex.mean())
df.sex.fillna(round_Values,inplace=True)
```

```python
df.sex.unique()
```

```python
df.isnull().sum()
```

# Replacing numerical values by median

```python
from statistics import median
df1 = df[['TSH','T3','TT4','T4U','FTI']]
df1 = df1.fillna(df1.median())
```

# Replacing target values by alphabets

```python
create =  df['Diagnosis'].str.split('([A-Za-z]+)', expand = True)
create = create[1]
df['Diagnosis'] = create.replace({None : 'Z'})
```

# Encoding target values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

cols = df.select_dtypes(include=['object'])
```

```python
for i in cols.columns:
    try:
        df[i] = le.fit_transform(df[i])
    except:
        continue
```

```python
df[['TSH', 'T3', 'TT4', 'T4U', 'FTI']] = df1[['TSH', 'T3', 'TT4', 'T4U', 'FTI']]
```

# Splitting into train and test datasets

```python
x = df.iloc[:, 1:26]
y = df.iloc[:,-1]
```

# Applying ML algorithms

## Decision Tree

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 50, test_size = 0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test)
```

```python
print("Train data accuracy:", accuracy_score(y_true = y_train, y_pred = clf.predict(x_train)))
print("Test data accuracy:", accuracy_score(y_true = y_test, y_pred = y_pred))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 50 )
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train)
```

# XGBoost

```python
from sklearn.model_selection import train_test_split
X6_train, X6_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 50)
```

```python
from xgboost import XGBRegressor
xg_reg = XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 5, gamma = 10, n_estimators = 10)
xg_reg.fit(X6_train, y_train)
y_pred = xg_reg.predict(X6_test)
```

```python
print("Train data accuracy:", xg_reg.score(X6_train, y_train))
print("Test data accuracy:", xg_reg.score(X6_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE: %f" % (rmse))
```
